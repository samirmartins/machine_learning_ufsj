```python
# Imports
import matplotlib.pyplot as plt

# Comentário: Exemplo de funcionalidades em Python

# Variáveis
nome = "Ana"
idade = 22

# Condicional
if idade >= 18:
    print(f"{nome} é maior de idade.")

# Lista e laço
notas = [7.5, 8.0, 9.2]

for i, nota in enumerate(notas):
    print(f"Nota {i+1}: {nota}")

# Função
def media(lista):
    return sum(lista) / len(lista)

print("Média:", media(notas))

# Gráfico simples
plt.plot([1, 2, 3], [7.5, 8.0, 9.2], marker='o')
plt.title("Notas")
plt.xlabel("Prova")
plt.ylabel("Nota")
plt.grid(True)
plt.show()
```
