# NumPy: arrays, operações vetorizadas e matrizes

Este notebook reúne exemplos curtos e executáveis sobre:
- criação de arrays e sequências numéricas;
- inspeção com `shape`, `dtype` e `size`;
- indexação e slicing;
- operações vetorizadas e funções úteis;
- matrizes e resolução de sistema linear.


```python
import numpy as np
```

## 1. Criando arrays e sequências

```python
notas = np.array([6.5, 7.0, 8.5, 9.0])
tempo = np.linspace(0, 10, 6)
pares = np.arange(0, 12, 2)

print("notas:", notas)
print("tempo:", tempo)
print("pares:", pares)
```

## 2. Shape, dtype e size

```python
medidas = np.array([[1.2, 1.5, 1.7],
                    [2.0, 2.2, 2.4]])

print("shape:", medidas.shape)
print("dtype:", medidas.dtype)
print("size:", medidas.size)
```

## 3. Indexação e slicing

```python
M = np.array([[10, 11, 12], 
              [20, 21, 22], 
              [30, 31, 32]])


#print("elemento [0, 1]:", M[0, 1])
#print("terceira coluna:", M[:, 2])
#print("submatriz:\n", M[1:, :2])

submatriz = M[1:, 1:(M.shape[1])]
print(submatriz)
```

## 4. Operações vetorizadas

```python
temperaturas_c = np.array([18, 21, 25, 27])

temperaturas_f = temperaturas_c * 9 / 5 + 32

aumentadas = temperaturas_c + 2

print("temperaturas em Celsius:", temperaturas_c)
print("temperaturas em Fahrenheit:", temperaturas_f)
print("temperaturas + 2:", aumentadas)
```

## 5. Funções úteis

```python
dados = np.array([4, 6, 8, 10, 12])

print("soma:", np.sum(dados))
print("media:", np.mean(dados))
print("maximo:", np.max(dados))
print("minimo:", np.min(dados))
```

## 6. Matrizes e álgebra linear

```python
A = np.array([[3, 1],
              [1, 2]])
B = np.array([[2, 0],
              [1, 4]])
b = np.array([9, 8])

print("A @ B =\n", A @ B)
print("solucao de A x = b:", np.linalg.solve(A, b))
```

## 7. Exemplo integrado

Vamos gerar um sinal, calcular estatísticas simples e reorganizar os dados em uma matriz.

```python
x = np.linspace(0, 2 * np.pi, 8)
sinal = np.sin(x)
matriz_sinal = sinal.reshape(2, 4)

print("sinal:", np.round(sinal, 3))
print("media do sinal:", np.mean(sinal))
print("matriz 2x4:\n", np.round(matriz_sinal, 3))
```
