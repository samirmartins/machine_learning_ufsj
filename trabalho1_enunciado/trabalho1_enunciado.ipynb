# Trabalho 1 (T1) — Análise de Dados Reais em Python

## 1. Finalidade do trabalho

- Consolidar, de forma prática e integrada, os conteúdos iniciais da disciplina, especialmente os itens **1 a 4** do conteúdo programático:
- Preparar o grupo para o segundo trabalho da disciplina
- Habituar ao uso de ferramentas essenciais em Machine Learning
- Desenvolver um estudo computacional completo, em Python, usando **dados reais** e produzindo uma análise clara, organizada, reprodutível e tecnicamente bem justificada.

---

## 2. Tema do trabalho

Cada grupo deverá escolher **um problema de interesse** e investigá-lo a partir de uma **base de dados reais**, obtida em **fonte confiável**, conforme discutido ao longo da disciplina. 

**IMPORTANTE:** o tema a ser trabalhado deverá ser acordado com o professor, mediante propostas apresentadas pelo grupo.

A proposta do trabalho não é, neste momento, construir um projeto avançado de modelagem preditiva ou classificatória, mas sim mostrar que o grupo consegue:

- obter dados reais;
- organizar e manipular esses dados em Python;
- realizar análise exploratória e estatística descritiva;
- produzir visualizações informativas;
- discutir o problema com linguagem adequada à disciplina;
- relacionar a base analisada com uma possível tarefa de aprendizado de máquinas (tema para o Trabalho 2).

---

## 3. Requisitos centrais: uso de dados reais e análises exploratórias

O trabalho deve obrigatoriamente utilizar **dados reais**.

Esses dados devem:

- ter origem em **fontes confiáveis**;
- ter sua fonte claramente indicada;
- estar adequados ao escopo da disciplina;
- permitir uma análise minimamente relevante.

Possíveis fontes serão discutidas e exemplificadas na disciplina.



---

## 4. O que o trabalho deve conter

O relatório (em formato .ipynb) deve contemplar, de forma organizada, os seguintes elementos.

### 4.1. Identificação
- título do trabalho;
- nomes dos integrantes do grupo;
- link do repositório GitHub do grupo.

### 4.2. Definição do problema
- descrição objetiva do tema;
- pergunta, problema ou motivação do estudo;
- justificativa breve da relevância do tema.

### 4.3. Fonte e obtenção dos dados
- o formato de obtenção dos dados (arquivo, API, portal de dados etc.);
- explicação de como os dados foram obtidos;
- observações sobre confiabilidade, contexto e limitações.
- uma justificativa da escolha da base e dos dados para o trabalho.

### 4.4. Leitura e organização da base em Python
- carregamento da base;
- descrição das variáveis;
- estruturação inicial dos dados;
- observações sobre formato, tipos e consistência.

### 4.5. Tratamento e preparação dos dados
- limpeza, quando necessária;
- tratamento de valores ausentes, inconsistências ou duplicidades;
- padronizações e transformações relevantes;
- justificativa das escolhas realizadas.

### 4.6. Análise exploratória e estatística
O grupo deve utilizar, quando fizer sentido para a base escolhida:

- medidas de tendência central;
- análise descritiva;
- distribuições;
- relações entre variáveis;
- discussão de correlação;
- comentários sobre causalidade e correlações espúrias, quando aplicável;
- situações envolvendo dependência, independência ou probabilidade condicional, quando couber.

### 4.7. Visualizações
O notebook deve apresentar um mínimo de 6 gráficos e visualizações distintas, úteis para a compreensão dos dados, com títulos, legendas e interpretação adequada.

**IMPORTANTE:** todos os gráficos construídos deverão ser analisados, e não somente apresentados.

### 4.8. Fechamento conceitual ligado a aprendizado de máquinas
Ao final, o grupo deve incluir uma discussão introdutória respondendo, sempre que possível:

- qual tipo de problema de aprendizado de máquinas poderia estar associado à base analisada;
- se o problema se aproxima mais de classificação, regressão ou outra formulação;
- quais dificuldades poderiam surgir em uma etapa posterior de modelagem;
- como os conceitos de sobreajuste, subajuste e viés-variância poderiam aparecer nesse contexto.

**IMPORTANTE:** Não é obrigatório que o grupo implemente, neste T1, modelos completos de aprendizado de máquinas. O foco é a preparação conceitual e analítica. Recomenda-se que o notebook alterne entre células de código e células em **markdown**, para melhor organização do trabalho.

## 6. Organização obrigatória no GitHub

O trabalho deverá estar organizado em um repositório no **GitHub**.

O repositório deve estar limpo, compreensível e estruturado de forma adequada, contendo, pelo menos:

- o notebook principal do trabalho;
- o arquivo `.yml` do ambiente;
- eventuais arquivos auxiliares;
- dados e instruções de acesso aos dados, quando couber;
- material de apresentação;
- arquivo README.md, contendo a descrição do repositório.

A ideia é que, ao longo da disciplina, seja possível constituir **um repositório da disciplina com todos os trabalhos desenvolvidos**.

## 7. Identificação da contribuição de cada integrante

É obrigatório indicar, de forma visível, **o que foi feito por cada integrante do grupo**.

Essa identificação deve ser feita **por cores** ao longo do notebook.

Exemplo de convenção:

- **Azul**: Aluno 1
- **Verde**: Aluno 2
- **Vermelho**: Aluno 3
- **Outra cor**: Aluno 4, se houver

Essa convenção deve aparecer logo no início do notebook.

Além disso:

- cada trecho relevante deve permitir identificar quem o produziu;
- a divisão de tarefas deve estar clara;
- todos os integrantes devem contribuir e todos devem apresentar.
